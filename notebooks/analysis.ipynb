# Campus Fiber Optic Network Optimization Analysis
This notebook demonstrates the step-by-step execution of Kruskal's Algorithm to find the Minimum Spanning Tree (MST) for a campus network, connecting various office switches to the central Core Switch with the lowest possible cabling cost.

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

## 1. Data Loading
We load the hypothetical campus network data from our CSV file. The edge weights represent the cabling costs in USD.

```python
data = pd.read_csv('../data/network_data.csv', sep=';')
data.head()
```

## 2. Graph Construction
We build the undirected graph where nodes are switches (buildings) and edges represent the possible fiber optic cable routes.

```python
G = nx.Graph()
for index, row in data.iterrows():
    G.add_edge(row['Source_Node'], row['Destination_Node'], weight=row['Cost_USD'])

print(f"Network created with {G.number_of_nodes()} nodes and {G.number_of_edges()} edges.")
```

## 3. Minimum Spanning Tree Execution (Kruskal's Algorithm)
Finding the optimal tree structure to connect all buildings without creating unnecessary loops, thereby minimizing the total capital expenditure (CAPEX).

```python
# Kruskal algoritması ile ağacı hesaplama
mst = nx.minimum_spanning_tree(G, weight='weight')
total_cost = mst.size(weight='weight')

print(f"Minimum Spanning Tree Total Cost: {total_cost} USD\n")
print("Selected Connections:")
for u, v, d in mst.edges(data=True):
    print(f"- {u} to {v}: {d['weight']} USD")
```
